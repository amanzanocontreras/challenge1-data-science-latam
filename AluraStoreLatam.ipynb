### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion_total = tienda["Precio"].sum()
facturacion_promedio = tienda["Precio"].mean()
top_facturacion = tienda.groupby("Producto")["Precio"].sum().sort_values(ascending=False).head(5)
```

```python
print(f"Facturación total estimada: ${facturacion_total:,.2f}")
print(f"Facturación promedio por venta: ${facturacion_promedio:,.2f}")
display(top_facturacion)
```

# 2. Ventas por categoría

```python
ventas_categoria = tienda.groupby("Categoría del Producto").agg({
    "Precio": ["count", "sum", "mean"]
})
ventas_categoria.columns = ["Número de ventas", "Total facturado", "Precio promedio"]
```

```python
display(ventas_categoria.sort_values(by="Total facturado", ascending=False))
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = tienda["Calificación"].mean()
calificacion_ciudad = tienda.groupby("Lugar de Compra")["Calificación"].mean().sort_values(ascending=False)
```

```python
print(f"Calificación promedio general: {calificacion_promedio:.2f}")
display(calificacion_ciudad)
```

# 4. Productos más y menos vendidos

```python
productos_vendidos = tienda["Producto"].value_counts()
```

```python
print("Productos más vendidos:")
display(productos_vendidos.head(5))
```

```python
print("Productos menos vendidos:")
display(productos_vendidos.tail(5))
```

# 5. Envío promedio por tienda

```python
envio_promedio = tienda.groupby("Lugar de Compra")["Costo de envío"].mean().sort_values()
```

```python
print("Costo promedio de envío por tienda o ciudad:")
display(envio_promedio)
```
